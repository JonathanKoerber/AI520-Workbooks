```python
import pandas as pd
from datasets import load_dataset

# a. Load the IMDb Movie Reviews dataset (train split)
dataset = load_dataset('imdb', split='train')

# b. Create a dataframe with specific columns
# We convert the Hugging Face dataset to a Pandas DataFrame first
df = pd.DataFrame(dataset)
df = df.rename(columns={'text': 'review_text', 'label': 'sentiment'})

# c. Map sentiment labels (0 -> ‘negative’, 1 -> ‘positive’)
sentiment_map = {0: 'negative', 1: 'positive'}
df['sentiment'] = df['sentiment'].map(sentiment_map)

# d. Save the processed dataset as movie_reviews.csv
# index=False prevents pandas from adding a column for the row numbers
df.to_csv('movie_reviews.csv', index=False)

print("File 'movie_reviews.csv' has been created successfully!")
print(df.head())
```

```python
import string
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # 1. make lowercase
    text = text.lower()
    # 2. remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)
    # 3. remove stop words
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Apply the function to the column and overwrite it
df['review_text'] = df['review_text'].apply(clean_text)

# Check the results
print(df['review_text'].head())
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# a. Download the necessary resources
# Note: In newer NLTK versions, 'punkt_tab' may be required alongside 'punkt'
nltk.download('punkt')
nltk.download('punkt_tab')

# b & c. Perform tokenization and store in new columns
# Sentence tokenization
df['sentences'] = df['review_text'].apply(sent_tokenize)

# Word tokenization
df['words'] = df['review_text'].apply(word_tokenize)

# Display the first few rows to verify
print(df[['sentences', 'words']].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# a. Create Bag-of-Words representation (Top 1000 features)
vectorizer = CountVectorizer(max_features=1000)
# We use the cleaned 'review_text' column from the previous steps
bow_matrix = vectorizer.fit_transform(df['review_text'])

# b. Calculate word frequencies
# Summing the counts for each column (word) across all rows
word_counts = bow_matrix.sum(axis=0).A1 
vocabulary = vectorizer.get_feature_names_out()

# c. Display the top 20 most frequent words
# Combine vocabulary and counts into a list of tuples
word_freq = list(zip(vocabulary, word_counts))

# Sort by frequency in descending order
top_20 = sorted(word_freq, key=lambda x: x[1], reverse=True)[:20]

# Print the results nicely
print(f"{'Word':<15} | {'Frequency':<10}")
print("-" * 25)
for word, freq in top_20:
    print(f"{word:<15} | {freq:<10}")
```

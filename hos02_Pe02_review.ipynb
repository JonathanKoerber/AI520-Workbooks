```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
print(nltk.__version__)
```

```python
text = """Natural language processing (NLP) is a subfield of artificial intelligence (AI) that focuses on the interaction between computers and humans through natural language. The ultimate goal of NLP is to enable computers to understand, interpret, and generate human language in a valuable way.
"""
sentence = sent_tokenize(text)
print("Tokenized sentences:")
print(sentence)
```

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(text)
print("Tokenized words:")
print(words)
```

```python
from nltk import FreqDist

fdisk = FreqDist(words)
print("Most common words")
print(fdisk.most_common(10))
```

```python
import string
nltk.download('stopwords')
from nltk.corpus import stopwords

text_no_punct = "".join([char for char in text if char not in string.punctuation])

stop_words = set(stopwords.words('english'))
words_no_stop = [word.lower() for word in word_tokenize(text_no_punct) if word.lower() not in stop_words]

print("Clean test tokens:")
print(words_no_stop)
```

```python
fdist_clean = FreqDist(words_no_stop)
print('Most common words after cleaning')
print(fdist_clean.most_common(10))
```

```python
import matplotlib.pyplot as plt

fdist_clean.plot(30, cumulative=False)
plt.title('Word Frequency Distribution')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

```python
import nltk
from nltk.corpus import reuters
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import numpy as np

# Download required NLTK data
nltk.download('reuters')
nltk.download('punkt')
```

```python
sample_texts = [
    "Natural language processing is fascinating. It helps computers understand human language.",
    "Machine learning models can process and analyze text data efficiently.",
    "Text representation is crucial for NLP tasks and applications."
]
```

```python
print("Number of documents:", len(sample_texts))
print("Total characters:", sum(len(text) for text in sample_texts ))
```

```python
# Create and configure CountVectorizer
bow_vectorizer = CountVectorizer(
    lowercase=True,
    token_pattern=r'\b\w+\b',
    min_df=1
)

# Generate BoW representation
X_bow = bow_vectorizer.fit_transform(sample_texts)

# Create a DataFrame for better visualization
bow_df = pd.DataFrame(
    X_bow.toarray(),
    columns=bow_vectorizer.get_feature_names_out()
)

print("\nBag-of-Words Representation:")
print(bow_df)
```

```python
tfidf_vectorizer = TfidfVectorizer(
    lowercase=True,
    token_pattern=r'\b\w+\b',
    min_df=1,
)
X_tfidf = tfidf_vectorizer.fit_transform(sample_texts)

tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Representation:")
print(tfidf_df)
```

```python
# Compare top terms by importance
def print_top_terms(vectorizer, matrix, n=5):
    sums = np.sum(matrix, axis=0)
    terms = vectorizer.get_feature_names_out()
    pairs = list(zip(terms, sums))
    pairs.sort(key=lambda x: x[1], reverse=True)
    
    print(f"\nTop {n} most important terms:")
    for term, score in pairs[:n]:
        print(f"{term}: {score:.3f}")

# Analyze both representations
print("\nBoW Analysis:")
print_top_terms(bow_vectorizer, X_bow.toarray())

print("\nTF-IDF Analysis:")
print_top_terms(tfidf_vectorizer, X_tfidf.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate similarity matrices
bow_similarity = cosine_similarity(X_bow)
tfidf_similarity = cosine_similarity(X_tfidf)

print("\nDocument Similarity Matrix (BoW):")
print(pd.DataFrame(bow_similarity, columns=['Doc1', 'Doc2', 'Doc3']))
```
